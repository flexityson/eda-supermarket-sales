**Goal:** Create 3 charts from the clean supermarket sales data.

---

## ✅ Step 1 — Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt

print('Libraries loaded ✅')
```

## ✅ Step 2 — Load the Clean Data

```python
df = pd.read_csv('clean_sales.csv')
print(f'Loaded: {df.shape[0]} rows, {df.shape[1]} columns')
df.head()
```

---
## 📊 Chart 1 — Bar Chart: Revenue by Category
> Which category makes the most money?

```python
# Prepare data
revenue_by_category = df.groupby('category')['total'].sum().sort_values(ascending=False)
print(revenue_by_category)
```

```python
# Draw the bar chart
plt.figure(figsize=(8, 5))
plt.bar(revenue_by_category.index, revenue_by_category.values, color='steelblue')

# Labels
plt.title('Total Revenue by Category', fontsize=14)
plt.xlabel('Category')
plt.ylabel('Total Revenue ($)')

plt.tight_layout()
plt.savefig('chart1_bar.png')
plt.show()
print('Chart 1 saved ✅')
```

```python
# 📝 YOUR OBSERVATION:
# Which category has the highest bar?
# Which has the lowest?
```

---
## 📈 Chart 2 — Line Chart: Sales Over Time
> How does total sales change day by day?

```python
# Prepare data
df['date'] = pd.to_datetime(df['date'])
sales_over_time = df.groupby('date')['total'].sum()
print(sales_over_time.head(10))
```

```python
# Draw the line chart
plt.figure(figsize=(10, 5))
plt.plot(sales_over_time.index, sales_over_time.values, color='green', marker='o', linewidth=2)

# Labels
plt.title('Daily Sales Over Time', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=45)

plt.tight_layout()
plt.savefig('chart2_line.png')
plt.show()
print('Chart 2 saved ✅')
```

```python
# 📝 YOUR OBSERVATION:
# Is there a day with a huge spike in sales?
# What might cause that spike?
```

---
## 🔵 Chart 3 — Scatter Plot: Price vs Rating
> Do more expensive products get better ratings?

```python
# Draw the scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(df['unit_price'], df['rating'], color='coral', alpha=0.6)

# Labels
plt.title('Unit Price vs Customer Rating', fontsize=14)
plt.xlabel('Unit Price ($)')
plt.ylabel('Rating')

plt.tight_layout()
plt.savefig('chart3_scatter.png')
plt.show()
print('Chart 3 saved ✅')
```

```python
# 📝 YOUR OBSERVATION:
# Do expensive products get higher ratings?
# Is there a clear pattern or is it random?
```
