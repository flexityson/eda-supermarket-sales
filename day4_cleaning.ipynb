## ✅ Step 1 — Import & Load

```python
import pandas as pd
import numpy as np

df = pd.read_csv('dirty_sales.csv')
print(f'Loaded: {df.shape[0]} rows, {df.shape[1]} columns')
```

## ✅ Step 3 — Find Missing Values
> Count how many missing values are in each column

```python
print('=== Missing Values ===')
print(df.isnull().sum())
print(f'\nTotal missing cells: {df.isnull().sum().sum()}')
```

```python
# 📝 YOUR OBSERVATION:
# Which column has the most missing values? Rating
# How many total missing cells? 20
```

## ✅ Step 4 — Find Duplicates
> Check if any rows appear more than once

```python
print(f'Duplicate rows: {df.duplicated().sum()}')
print('\nThe duplicate rows are:')
df[df.duplicated()]
```

```python
# 📝 YOUR OBSERVATION:
# How many duplicates did you find? 3
# Which order_ids appear twice? None
```

## ✅ Step 5 — Clean Step 1: Remove Duplicates

```python
df_clean = df.drop_duplicates()
print(f'Before: {df.shape[0]} rows')
print(f'After : {df_clean.shape[0]} rows')
print(f'Removed: {df.shape[0] - df_clean.shape[0]} duplicate rows')
```

## ✅ Step 6 — Clean Step 2: Fix Missing Values
> Different columns need different strategies

```python
# Numeric columns → fill with mean (average)
df_clean['customer_age']  = df_clean['customer_age'].fillna(df_clean['customer_age'].mean())
df_clean['rating']        = df_clean['rating'].fillna(df_clean['rating'].mean())
df_clean['unit_price']    = df_clean['unit_price'].fillna(df_clean['unit_price'].mean())

print('Numeric columns fixed ✅')
```

```python
# Text columns → fill with 'Unknown'
df_clean['customer_gender'] = df_clean['customer_gender'].fillna('Unknown')
df_clean['payment_method']  = df_clean['payment_method'].fillna('Unknown')

print('Text columns fixed ✅')
```

## ✅ Step 7 — Verify the Clean Data

```python
print('=== Missing Values After Cleaning ===')
print(df_clean.isnull().sum())
print(f'\nDuplicates remaining: {df_clean.duplicated().sum()}')
print(f'\nFinal shape: {df_clean.shape}')
```

```python
# 📝 YOUR OBSERVATION:
# Are there any missing values left? 0
# How many rows in the clean dataset? 50
```

## ✅ Step 8 — Save the Clean Dataset

```python
df_clean.to_csv('clean_sales.csv', index=False)
print('✅ Clean dataset saved as clean_sales.csv')
```
