## ✅ Step 6 — Answer These 3 Questions (Your Turn!)

### ❓ Q1: Which category generates the most total revenue?

```python
# HINT: groupby('category') then sum 'total', then sort
df.groupby('category')['total'].sum().sort_values(ascending=False)
#Electronics
```

### ❓ Q2: Which city has the highest average customer rating?

```python
# HINT: groupby('city') then mean 'rating'
df.groupby('city')['rating'].mean().sort_values(ascending=False)
#Chiang Mai
```

### ❓ Q3: What is the most popular payment method?

```python
# HINT: df['payment_method'].value_counts()
df['payment_method'].value_counts().sort_values(ascending=False)
#Credit Card 
#E-wallet
```

## ✅ Step 7 — Bonus Challenges 🔥
